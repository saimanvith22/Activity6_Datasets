Data Visualization with Seaborn and Matplotlib:-


Histogram:- The style of the plots is set to "whitegrid" using sns.set_style(), which provides a nice background for the graphs.
A loop iterates through each feature to create a histogram for each feature. These plots show the distribution of data points across different values of each feature.i.e. Sepal_length, Sepal_width, Petal_length, Petal_width.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Setting the style of the plots
sns.set_style("whitegrid")

# Assuming your dataset has similar columns to the Iris dataset
features = df28.columns[:-1]  # Excluding the species/target column
for feature in features:
    plt.figure(figsize=(7, 4))
    sns.histplot(df28[feature], kde=True, bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

Pairplot:- A pairplot is created using sns.pairplot(), plotting pairwise relationships in the dataset. The hue='species' parameter colors points by the species column, which is assumed to be the target variable, facilitating the visualization of how species are distributed with respect to the features.

```python
# Pairplot to visualize pairwise relationships in the dataset.
sns.pairplot(df28, hue='species', height=2.5)
plt.savefig('new.png')
plt.show()
```

Heatmap:- A heatmap of the correlation matrix (df28.corr()) visualizes the correlation coefficients between pairs of features, providing insights into how features relate to each other.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Correlation matrix heatmap
plt.figure(figsize=(8, 6))
# as the correleation can be done only for numerical values hence rectified the error by converting the data type
numerical_df = df28.select_dtypes(include=['number'])
sns.heatmap(numerical_df.corr(), annot=True, fmt=".2f", cmap='Blues')
#sns.heatmap(df28.corr(), annot=True, fmt=".2f", cmap='Blues')
plt.title('Correlation Matrix')
plt.show()
```

Data Visualization using Altair

Two plots are defined: a scatter plot and a bar plot. Both plots are encoded with petal length on the x-axis and petal width on the y-axis, colored by species.

The scatter plot (mark_point()) visualizes the relationship between petal length and petal width, with each point representing an observation in the dataset colored by species.

The bar plot (mark_bar()) displays a similar relationship using bars, providing a different visual perspective on the data.

```python
import altair as alt
import pandas as pd
scatterplot= alt.Chart(df28).mark_point().encode(
    x=('bill_length_mm'), #assigning the petal length on the x-axis
    y=('bill_depth_mm'), #assigning the petal width on the Y-axis
    color='species', #assigning color to species

)
barPlot=alt.Chart(df28).mark_bar().encode(
     x=('bill_length_mm'), #assign the petal length on the x-axis
    y=('bill_depth_mm'),#assign the petal width on Y-axis
    color='species' #assigning color to species
).interactive()

alt.hconcat(scatterplot, barPlot) #concatenate two plots
```

The above Visualization did concatenation of the 2 Plots.

```python
# Start coding here
import altair as alt

# Box Plot for Bill Length by Species
box_plot = alt.Chart(df28).mark_boxplot().encode(
    x='species:N',
    y='bill_length_mm:Q',
    color='species:N',
    tooltip=['species:N', 'bill_length_mm:Q']
).properties(
    title='Box Plot of Bill Length by Species'
)

# Heatmap showing correlation between bill_length_mm and bill_depth_mm
heatmap = alt.Chart(df28).mark_rect().encode(
    x='bill_length_mm:Q',
    y='bill_depth_mm:Q',
    color='count():Q',
    tooltip=['bill_length_mm:Q', 'bill_depth_mm:Q', 'count():Q']
).properties(
    title='Heatmap of Bill Length and Bill Depth'
).transform_aggregate(
    count='count()',
    groupby=['bill_length_mm', 'bill_depth_mm']
)

# Combine the box plot and heatmap with a shared legend
combined_plot = alt.vconcat(box_plot, heatmap).resolve_scale(
    color='independent'  # Ensures the color scale for species is independent
)

# Display the combined plot
combined_plot

```

Altair is built on top of Vega and Vega-Lite visualization grammars, providing a powerful, yet simple, way to construct a wide range of visualizations. Here, In this tutorial we will be working on simple plotting along with interactive Visualizations which will enhance our thinking of what the data represents.

alt.Chart(df28):- This initializes a chart object with df28 as the data source.

.mark_point(shape='circle'): Specifies that the chart will use circles to represent data points.

.encode(): This is used to map data columns to visuals or encoding them such as the x-axis, y-axis.

x=alt.X('Entity'): Maps the "Entity" to the x-axis.

y=alt.Y('Cellular Subscription'): Maps the "Cellular Subscription" to the y-axis.

```python
# plotting points on X and Y axis for with our desired column names
alt.Chart(df28).mark_point(shape='circle').encode(
    x=alt.X('make'),
    y=alt.Y('horsepower')
)
```

```python
Code_details = df28.torque.unique()
options = Code_details.tolist()
labels = [str(option) + ' ' for option in options]

input_dropdown = alt.binding_radio(
    options=options + [None],
    labels=labels + ['All'],
    name='torque:'
)
selection = alt.selection_single(
    fields=['torque'],
    bind=input_dropdown
)

chart_interactive = alt.Chart(df28).mark_point().encode(
    x= 'horsepower',
    y='make',
    color=alt.condition(selection,alt.Color('highway_mpg:Q'),alt.value('lightgrey'),legend = None),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
    ).add_selection(selection)


chart_interactive
```

Based on the above graph you can observe that we have a single selction chart where you can select any country and the graph only shows the data of that country just like how filters works.
alt.binding_radio(): It Creates a radio button for selecting different codes. It also includes an "All" option to remove the selection filter.

alt.selection_single(): It Creates a single selection interaction that allows users to select one option from the dropdown.

The chart_interactive constructs an interactive chart where data point's color and opacity will change based on the selection from the dropdown. The color and opacity are conditionally set based on whether a selection is made.

```python
alt.Chart(df28).mark_point().encode(
    x=alt.X('torque'),
    y=alt.Y('horsepower'),
    color='make', tooltip=['highway_mpg']
)
```

```python
# Display 2-3 visualization of your choice

import altair as alt


pie_data = data[['make', 'horsepower']].drop_duplicates()

# Create Pie Chart
pie_chart = alt.Chart(pie_data).mark_arc().encode(
    theta='horsepower:Q',
    color='make:N',
    tooltip=['make:N', 'horsepower:Q']
).properties(
    title='Pie Chart of Horsepower by Vehicle Make'
)

# Create Donut Chart
donut_chart = alt.Chart(pie_data).mark_arc(innerRadius=50).encode(
    theta='horsepower:Q',
    color='make:N',
    tooltip=['make:N', 'horsepower:Q']
).properties(
    title='Donut Chart of Horsepower by Vehicle Make'
)

# Step 2: Create a scatter plot with triangle shapes
scatter_plot = alt.Chart(data).mark_point(shape='triangle').encode(
    x=alt.X('make:N', title='Vehicle Make'),
    y=alt.Y('horsepower:Q', title='Horsepower'),
    color='make:N',  # Color points based on vehicle make
    tooltip=['make:N', 'horsepower:Q']
).properties(
    title='Scatter Plot of Horsepower by Vehicle Make'
)

# Concatenate pie and donut charts horizontally
combined_charts = alt.hconcat(pie_chart, donut_chart).resolve_scale(
    color='independent'
)


combined_charts & scatter_plot



```
